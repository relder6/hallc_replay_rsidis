```python
import pandas as pd
```

```python
df = pd.read_csv("updated_parsed_runlist_110425.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df_sidis = df["run"][
    ((df["run_type"] == "PI-SIDIS") | (df["run_type"] == "PI+SIDIS")) 
]
```

```python
len(df_sidis)
```

```python
df_sidis.to_csv("sidis_runlist_pass0p1.txt", index=False, header=False)
```

```python
df_heep = df["run"][
    ((df["run_type"] == "HEEP")) 
]
```

```python
df_heep.to_csv("heep_runlist_pass0p1.txt", index=False, header=False)
```

```python
df_hms = df["run"][
    ((df["run_type"] == "HMSDIS") | (df["run_type"] == "OPTICS") | ( (df["run_type"] == "HEE") & (df["ps4"] != -1) )) 
]
```

```python
df_hms.to_csv("hms_runlist_pass0p1.txt", index=False, header=False)
```

```python
df_shms = df["run"][
    ((df["run_type"] == "SHMSDIS") | (df["run_type"] == "OPTICS") | ( (df["run_type"] == "HEE") & (df["ps2"] != -1) )) 
]
```

```python
df_shms.to_csv("shms_runlist_pass0p1.txt", index=False, header=False)
```
